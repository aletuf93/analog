### Import packages

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_wine
from sklearn.manifold import MDS

import sys; sys.path.insert(0, '..') #add the above level with the package
from analogistics import reduce as dr
from analogistics import explore


import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

### Import the dataset

```python
#Load the dataset
data = load_wine()

# define X dataframe
X = data.data
X_labels = data.feature_names
X = pd.DataFrame(X,columns=X_labels)

# define y dataframe
y = data.target
y = pd.DataFrame(y,columns=['target'])
```

### Feature extraction
These methods are used to transform the dataset, maintaining the greater amount of intormation possible.

```python
#Extract features using Principal Component Analysis (PCA)

#analyse the information content of each principal component
X_res, out_fig_dict, output_df = dr.PCAplot(n_comp=len(X.columns), df=X, diagnose=True)
# out_fig_dict['PCA_varianceExplanation']





```

```python
# plot the dataset reduced into only two components
# the axis identify the percentage of variance explained by each PC
X_res, out_fig_dict, output_df = dr.PCAplot(n_comp=2, df=X, diagnose=True)
# out_fig_dict['PCA_plot']
```

```python
#Extract features using Multi-dimensional scaling

mds = MDS(random_state=0)
X_transform = mds.fit_transform(X)

plt.scatter(X_transform[:,0], X_transform[:,1])
plt.title('Embedding in 2D')
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()

```

### Feature selection
These methods are used to select the features that retains the greatest amount of information.

```python
#Define the correlation matrix of the inpyt dataframe
output_figures = explore.correlationMatrix(X,annotationCell=True)
output_figures['CorrelationMatrix']
```

```python
# Select features by correlation
corrThreshold = 0.3
Z, out_fig_dict = dr.selectByCorrelation(X,y,corrThreshold=corrThreshold, diagnose=True)
out_fig_dict['CorrelationChart']


```

```python
#Select features using variance
varThreshold = 0.3
Z, out_fig_dict = dr.selectByVariance(X, perc=varThreshold,diagnose=True)
out_fig_dict.keys()
out_fig_dict['VarianceChart']
```

```python
#Select features using Lasso
Z, out_fig_dict = dr.selectByLassoL1(X,y,value=0.5,diagnose=True)

out_fig_dict.keys()
out_fig_dict['LassoChart']
```

```python
# lasso path
out_fig_dict = dr.lassoPath(X,y)
out_fig_dict['LassoPath']
```

```python
#Select features using a tree
Z = dr.selectByTree(X,y)
```

```python
#Select features using forward stepwise selection
X_res, out_fig_dict, output_df = dr.selectByForwardStepwiseSelection(X,y,n_feat=5)
out_fig_dict.keys()
```
